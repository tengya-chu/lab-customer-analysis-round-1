```python
# Read the three files into python as dataframes
import pandas as pd
import numpy as np
file1 = pd.read_csv('./files_for_lab/csv_files/file1.csv')
file2 = pd.read_csv('./files_for_lab/csv_files/file2.csv')
file3 = pd.read_csv('./files_for_lab/csv_files/file3.csv')
```

```python
# Show the DataFrame's shape.
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
# Standardize header names.
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
cols = []
for i in range(len(file1.columns)):
    cols.append(file1.columns[i].lower())
file1.columns = cols
```

```python
cols = []
for i in range(len(file2.columns)):
    cols.append(file2.columns[i].lower())
file2.columns = cols
```

```python
cols = []
for i in range(len(file3.columns)):
    cols.append(file3.columns[i].lower())
file3.columns = cols
```

```python
file1 = file1.rename(columns={ 'st':'state'})
file2 = file2.rename(columns={ 'st':'state'})
```

```python
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
file1 = file1[['customer', 'state', 'customer lifetime value', 'education', 'gender',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'total claim amount', 'vehicle class']]
file2 = file2[['customer', 'state', 'customer lifetime value', 'education', 'gender',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'total claim amount', 'vehicle class']]
file3 = file3[['customer', 'state', 'customer lifetime value', 'education', 'gender',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'total claim amount', 'vehicle class']]
```

```python
np.all(file1.columns == file2.columns)
```

```python
# Concatenate the three dataframes
data = pd.concat([file1, file2, file3])
```

```python
data
```

```python
# Which columns are numerical? Which columns are categorical?
data.dtypes
```

Perform the data cleaning operations mentioned so far in class

```python
# Delete the column education and the number of open complaints from the dataframe.
data = data.drop(['education', 'number of open complaints'], axis=1)
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
data['customer lifetime value'] = data['customer lifetime value'].str.rstrip("%").astype(float)/100
```

```python
data
```

```python
# Check for duplicate rows in the data and remove if any.
data = data.drop_duplicates()
```

```python
data
```

```python
# Filter out the data for customers who have an income of 0 or less.
data[(data['income']<=0)]
```
